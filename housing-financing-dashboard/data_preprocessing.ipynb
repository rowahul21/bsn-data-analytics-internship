```python
#Ambil kcs induk aja
df_kcs = df[df['Cabang'].str.contains('KCS', case=False, na=False)].copy()
```

```python
#Ambil KCS unik & buat kode baru (36 sesuai data)
unique_kcs = sorted(df_kcs['wilayah'].unique())

# bikin kode unik (misalnya 100–230)
import numpy as np
kode_baru = np.linspace(100, 230, len(unique_kcs)).astype(int)

mapping_kcs = dict(zip(unique_kcs, kode_baru))
```

```python
#Fungsi mapping KCPS → KCS
def map_ke_kcs(wilayah):
    for kcs_wilayah in mapping_kcs:
        if kcs_wilayah in wilayah:
            return mapping_kcs[kcs_wilayah]
    return None
```

```python
#Apply ke seluruh data
def assign_kode(row):
    wilayah = row['wilayah']
    
    if 'KCS' in row['Cabang'].upper():
        return mapping_kcs.get(wilayah)
    else:
        return map_ke_kcs(wilayah)

df['Kode_KCS'] = df.apply(assign_kode, axis=1)
```

```python
#definisi range + probabilitas

np.random.seed(42)

# fungsi generate tenor per baris
def generate_tenor():
    r = np.random.rand()
    
    if r < 0.02:
        return np.random.randint(12, 60)     # 1–5 tahun
    elif r < 0.05:
        return np.random.randint(60, 108)    # 6–9 tahun
    elif r < 0.20:
        return np.random.randint(108, 180)   # 10–15 tahun
    elif r < 0.80:
        return np.random.randint(180, 240)   # 15–20 tahun ✅ PALING BANYAK
    else:
        return np.random.randint(240, 300)   # >20 tahun
```

```python
#apply ke kolom tenor
df['Tenor'] = [generate_tenor() for _ in range(len(df))]
```

```python
#14. Tanggal
# menentukan distribusi per tahun
unit_per_tahun = {
    2021: 20000,
    2022: 35000,
    2023: 44000,
    2024: 45000,
    2025: 70000
}
```

```python
#Generate kolom Tgl_Akad berdasarkan tahun
np.random.seed(42)

data_list = []

unit_counter = 1

for tahun, jumlah in unit_per_tahun.items():
    for _ in range(jumlah):
        
        # random tanggal dalam tahun tersebut
        bulan = np.random.randint(1, 13)
        hari = np.random.randint(1, 28)  # biar aman
        
        tgl_akad = pd.Timestamp(year=tahun, month=bulan, day=hari)
        
        # tenor random (10–30 tahun)
        tenor_tahun = np.random.choice([10, 15, 20, 25, 30])
        tgl_jatuh_tempo = tgl_akad + pd.DateOffset(years=tenor_tahun)
        
        data_list.append([
            unit_counter,
            tgl_akad,
            tgl_jatuh_tempo
        ])
        
        unit_counter += 1

# buat dataframe baru
df_time = pd.DataFrame(data_list, columns=['Unit', 'Tgl_Akad', 'Tgl_Jatuh_Tempo'])
```

```python
#gabungkan ke df utama
df[['Unit', 'Tgl_Akad', 'Tgl_Jatuh_Tempo']] = df_time[['Unit', 'Tgl_Akad', 'Tgl_Jatuh_Tempo']]
```
